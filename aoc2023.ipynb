```python
# Day 1

import re
with open("day1input.txt") as file:
    lines = [line.rstrip() for line in file]

def calibration_values(lines: str):
    num_list = []
    for line in lines:
        nums_in_line = re.findall(r'\d', line)
        num_list.append(int(nums_in_line[0] + nums_in_line[-1]))
    return sum(num_list), num_list

print(calibration_values(lines))
```

```python
word_digit_pairs = [
    ('one', 'o1e'),
    ('two', 't2o'),
    ('three', 't3e'),
    ('four', 'f4r'),
    ('five', 'f5e'),
    ('six', 's6x'),
    ('seven', 's7s'),
    ('eight', 'e8t'),
    ('nine', 'n9e')
]
lines_replaced = lines
for word, digit in word_digit_pairs:
    lines_replaced = [x.replace(word, digit) for x in lines_replaced]
print(calibration_values(lines_replaced))
```

```python
# Day 2
games = {}
with open("day2_input.txt") as file:
    for line in file:
        game_result = {}
        game_line = line.rstrip().split(':')
        game_id = int(game_line[0].replace('Game ', ''))
        game_filtered = game_line[1].split(';')
        output_list = [
            {'blue': 0, 'green': 0, 'red': 0, **{color: int(quantity) for quantity, color in (pair.strip().split(' ') for pair in item.split(','))}}
            for item in game_filtered
        ]
        games[game_id] = output_list
flagged_games = []            
for game_id, game in games.items():
    excluded_game = False
    for g in game:
        if g['blue'] > 14 or g['green'] > 13 or g['red'] > 12:
            excluded_game = True
    if not excluded_game:
        flagged_games.append(game_id)

print(sum(flagged_games))
```

```python
powers = []
for _,game in games.items():
    min_blue_list = []
    min_red_list = []
    min_green_list = []
    for g in game:
        min_blue_list.append(g['blue'])
        min_red_list.append(g['red'])
        min_green_list.append(g['green'])
    min_blue = max(min_blue_list)
    min_red = max(min_red_list)
    min_green = max(min_green_list)
    powers.append(min_blue * min_red * min_green)

print(sum(powers))
```
